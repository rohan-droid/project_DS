```python
#6.Identify significant variables by performing a statistical analysis using p-values and (Optional)
#6.Identify significant variables by performing a statistical analysis using chi-square values (Optional)
# h0: the average response time is the same
# ha: the average response time is different
# if alpha is 5% or 0.05
# p value -0.0000000000
#p < alpha, so we will reject the h0
#the average response time across types of complaints is different
# anova will be for a continous variable
```

```python
# look at top 5 complaint types and find if the average response time is same or different
df1=dfsr[dfsr['Complaint Type']=='Blocked Driveway']['response_hours'].dropna()
df2=dfsr[dfsr['Complaint Type']=='Illegal Parking']['response_hours'].dropna()
df3=dfsr[dfsr['Complaint Type']=='Noise - Street/Sidewalk']['response_hours'].dropna()
df4=dfsr[dfsr['Complaint Type']=='Noise - Commercial']['response_hours'].dropna()
df5=dfsr[dfsr['Complaint Type']=='Derelict Vehicle']['response_hours'].dropna()
```

```python
stats.f_oneway(df1,df2,df3,df4,df5)
```
